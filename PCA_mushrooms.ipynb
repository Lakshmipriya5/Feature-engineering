```python
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
```

```python
for colm in data.columns:
    data[colm]=lb.fit_transform(data[colm])
```

```python
data.columns
```

```python
X=data.iloc[:,1:23]
y=data.iloc[:,0]
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X=sc.fit_transform(X)
```

```python
pca=PCA()
pca.fit_transform(X)
pca_variance=pca.explained_variance_

```

```python
plt.figure(figsize=(8,6))
plt.bar(range(22),pca_variance,alpha=0.5,align='center',label='individual variance')
plt.title('PCA visualization')
plt.xlabel('variance ratio')
plt.ylabel('principle components')
```

```python
pca2=PCA(n_components=17)
pca2.fit(X)
x_data=pca2.transform(X)
```

```python
plt.figure(figsize=(8,6))
plt.scatter(x_data[:,0], x_data[:,5], c=data['class'])
plt.show()
```

```python
pca3=PCA(n_components=2)
pca3.fit(X)
x_data3=pca3.transform(X)
```

```python
plt.figure(figsize=(8,6))
plt.scatter(x_data3[:,0], x_data3[:,1], c=data['class'])
plt.show()
```
